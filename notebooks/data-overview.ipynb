```python
TRAIN_DIR = '../dataset/train'
```

```python
import os
```

```python
classes = os.listdir(TRAIN_DIR)
len(classes)
```

```python
from operator import itemgetter
```

```python
total_images = 0
```

```python
most_populated_classes = {}

for training_class in classes:
    images = len(os.listdir(f"{TRAIN_DIR}/{training_class}"))
    total_images += images
    most_populated_classes[training_class] = images
    
most_populated_classes = dict(sorted(most_populated_classes.items(), key=itemgetter(1), reverse=True))
```

```python
total_images
```

```python
classes = list(most_populated_classes.keys())[:10]
classes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
from random import choice
```

```python
plt.figure(figsize=(20,10))

for index, training_class in enumerate(classes):
    available_images = os.listdir(f"{TRAIN_DIR}/{training_class}")
    img = Image.open(f"{TRAIN_DIR}/{training_class}/{choice(available_images)}")
    plt.subplot(2, 5, index+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(img)
    plt.xlabel(training_class)

plt.savefig('../images/data.jpg', bbox_inches='tight', pad_inches=0)
plt.show()
```
